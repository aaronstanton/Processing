```julia
using Seismic,SeismicPlotting
```

```julia
param = ["adj"=>"y",
    "vp"=>"../prep/vp", "vs"=>"../prep/vs", "wav"=>"../prep/wav", 
    "angpx"=>"../prep/angpx","angpy"=>"../prep/angpy",
    "angsx"=>"../prep/angsx","angsy"=>"../prep/angsy",
    "pade_flag"=>"y",
    "sz"=>10, "gz"=>550,
    "fmin"=>1., "fmax"=>50.,
    "padt"=>2, "padx"=>1,
    "nangx"=>71, "oangx"=>-70, "dangx"=>2,
    "nangy"=>1, "oangy"=>0, "dangy"=>2,
    "nhx"=>827, "ohx"=>-3304, "dhx"=>8,
    "nhy"=>1, "ohy"=>0, "dhy"=>8,
    "omp"=>8,"verbose"=>"y"];
param["sx"] = [5100.0:100.0:5100.0];
param["sy"] = [5100.0:100.0:5100.0]*0;
ShotProfileEWEM("../prep/ux","../prep/uy","../prep/uz","mpp","mps","msp","mss",param)
```

```julia
SeisHeaderInfo("mpp")
```

```julia
SeisWindow("mpp","mpp_mute",["ang"],[-40],[40])
SeisStack("mpp_mute","mpp_stack",["key"=>["imx","imy"]]);
```

```julia
mpp_stack,h = SeisRead("mpp_stack");
plotpar = ["style"=>"color","vmin"=>-1e2,"vmax"=>1e2,"aspect"=>"1.5","xlabel"=>"X","xunits"=>"(m)","ox"=>0.,"dx"=>8.,"ylabel"=>"Depth","yunits"=>"(m)","oy"=>0.,"dy"=>4,"cmap"=>"Greys"];
SeisPlot(mpp_stack,["vmin"=>-1e3,"vmax"=>1e3,"cmap"=>"PuOr"]);
```
