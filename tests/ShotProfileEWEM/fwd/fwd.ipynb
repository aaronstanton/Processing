```julia
using Seismic
```

```julia
param = ["adj"=>"n",
         "vp"=>"../prep/vp", "vs"=>"../prep/vs", "wav"=>"../prep/wav", 
         "pade_flag"=>"y",
         "sz"=>10, "gz"=>550,
         "fmin"=>0., "fmax"=>50.,
         "padt"=>2, "padx"=>2,
         "nang"=>1, "oang"=>0, "dang"=>2,
         "naz"=>1, "oaz"=>0, "daz"=>2,
         "nhx"=>827, "ohx"=>-3304, "dhx"=>8,
         "nhy"=>1, "ohy"=>0, "dhy"=>8,
         "omp"=>8,"verbose"=>"y"];
param["sx"] = [3600.0:100.0:3600.0];
param["sy"] = [3600.0:100.0:3600.0]*0;
ShotProfileEWEM("ux_fwd","uy_fwd","uz_fwd","../adj/mpp","../adj/mps","../adj/msp","../adj/mss",param);
```

```julia
ux_fwd,h = SeisRead("ux_fwd");
uz_fwd,h = SeisRead("uz_fwd");
plotpar = ["style"=>"color","vmin"=>-3e-9,"vmax"=>3e-9,"aspect"=>"auto","xlabel"=>"Offset","xunits"=>"(m)","ox"=>-3304.,"dx"=>8.,"ylabel"=>"Time","yunits"=>"(s)","oy"=>0.,"dy"=>0.002,"wbox"=>8,"hbox"=>8,"cmap"=>"Greys"];
Seismic.subplot(121)
SeisPlot(ux_fwd,plotpar);Seismic.xticks(vec([-3000:1500:3000]));
Seismic.subplot(122)
SeisPlot(uz_fwd,plotpar);Seismic.xticks(vec([-3000:1500:3000]));
```

```julia
fmax=60;
dk = 1/8./size(ux_fwd,2)
kmin = -dk*size(ux_fwd,2)/2
kmax =  dk*size(ux_fwd,2)/2
df = 1/0.002/size(ux_fwd,1)
FMAX = df*size(ux_fwd,1)/2 
nf = convert(Int32,floor((size(ux_fwd,1)/2)*fmax/FMAX))
Ux = abs(fftshift(fft(ux_fwd)))
Ux = Ux[end/2:end/2+nf,:]
Uz = abs(fftshift(fft(uz_fwd)))
Uz = Uz[end/2:end/2+nf,:]
Seismic.subplot(121)
Seismic.imshow(Ux,vmin=0,vmax=1e-5,extent=[kmin,kmax,fmax,0],aspect="auto")
Seismic.subplot(122)
Seismic.imshow(Uz,vmin=0,vmax=1e-5,extent=[kmin,kmax,fmax,0],aspect="auto")
```
