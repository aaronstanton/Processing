```julia
using Seismic
```

```julia
param = ["adj"=>"n",
         "vel"=>"../prep/vp","wav"=>"../prep/wav", 
         "pade_flag"=>"n",
         "sz"=>10, "gz"=>550,
         "fmin"=>0., "fmax"=>50.,
         "padt"=>2, "padx"=>2,
         "nang"=>1, "oang"=>0, "dang"=>2,
         "naz"=>1, "oaz"=>0, "daz"=>2,
         "nhx"=>827, "ohx"=>-3304, "dhx"=>8,
         "nhy"=>1, "ohy"=>0, "dhy"=>8,
         "omp"=>8,"verbose"=>"y"];
param["sx"] = [3600.0:100.0:3600.0];
param["sy"] = [3600.0:100.0:3600.0]*0;
ShotProfileWEM("../adj/mpp","uz_fwd",param);
```

```julia
uz_fwd,h = SeisRead("uz_fwd");
plotpar = ["style"=>"color","vmin"=>-4e-2,"vmax"=>4e-2,"aspect"=>"auto","xlabel"=>"Offset","xunits"=>"(m)","ox"=>-3304.,"dx"=>8.,"ylabel"=>"Time","yunits"=>"(s)","oy"=>0.,"dy"=>0.002,"wbox"=>8,"hbox"=>8,"cmap"=>"Greys"];
Seismic.subplot(121)
SeisPlot(uz_fwd,plotpar);Seismic.xticks(vec([-3000:1500:3000]));
```

```julia
fmax=60;
dk = 1/8./size(uz_fwd,2)
kmin = -dk*size(uz_fwd,2)/2
kmax =  dk*size(uz_fwd,2)/2
df = 1/0.002/size(uz_fwd,1)
FMAX = df*size(uz_fwd,1)/2 
nf = convert(Int32,floor((size(uz_fwd,1)/2)*fmax/FMAX))
Uz = abs(fftshift(fft(uz_fwd)))
Uz = Uz[end/2:end/2+nf,:]
Seismic.subplot(121)
Seismic.imshow(Uz,vmin=0,vmax=200,extent=[kmin,kmax,fmax,0],aspect="auto")
```
