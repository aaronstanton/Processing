```julia
using Seismic
```

```julia
SegyToSeis("/Users/astanton/Seismic_SAIG/data/syn_cmp.su","syn_cmp","su",false);
d,headers = SeisRead("syn_cmp");
h = Float64[];
for ix = 1 : length(headers)
    push!(h,headers[ix].h);
end 
```

```julia
plotpar = ["style"=>"wiggles",
           "vmin"=>-2,"vmax"=>2,
           "aspect"=>"auto",
    "xlabel"=>"Offset","xunits"=>"(meters)","ox"=>0,"dx"=>headers[2].h-headers[1].h,
           "ylabel"=>"Time","yunits"=>"(seconds)","oy"=>0,"dy"=>headers[1].d1,
           "cmap"=>"gray",
           "title"=>"syn_cmp.su"];
Seismic.subplot(121);
SeisPlot(d,plotpar);
```

```julia
S,tau,v = SeisSemblance(d,["offset"=>h,"dt"=>headers[1].d1]);
plotpar_semblance = ["style"=>"color",
           "vmin"=>0,"vmax"=>0.5,
           "aspect"=>"auto",
           "xlabel"=>"RMS Velocity","xunits"=>"","ox"=>minimum(v),"dx"=>v[2]-v[1],
           "ylabel"=>"Tau","yunits"=>"(seconds)","oy"=>tau[1],"dy"=>tau[2]-tau[1],
           "cmap"=>"jet",
           "title"=>"Semblance: syn_cmp.su"];
Seismic.subplot(121);
SeisPlot(S,plotpar_semblance);
```

```julia
tnmo = [0.5 1.25 1.75];
vnmo = [2000 2300 2500];
dnmo = SeisNMO(d,["tnmo"=>tnmo,"vnmo"=>vnmo,"offset"=>h,"dt"=>headers[1].d1]);
Seismic.subplot(121);
SeisPlot(dnmo,plotpar);
```
