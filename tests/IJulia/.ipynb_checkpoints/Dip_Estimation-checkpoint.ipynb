```julia
using Seismic
```

```julia
d1 = SeisLinearEvents(["tau1"=>0.2,"nx1"=>500,"v1"=>8500,"f0"=>20]);
d1 = d1[:,:];
d2 = SeisLinearEvents(["tau1"=>1.2,"nx1"=>500,"v1"=>-1000,"f0"=>20]);
d2 = d2[:,:];
d = d1 + d2 + 0.02*randn(size(d1));
```

```julia
coh,pp,res = SeisPWD(d,["w1"=>50,"w2"=>50]);
```

```julia
SeisPlot(d,["vmin"=>-1,"vmax"=>1]);
Seismic.hold("true");
L = 1;
pp_x = cos((pp-90)*pi/180)*L
pp_y = sin((pp-90)*pi/180)*L
Seismic.quiver([0:30:500-1],[0:30:500-1],
pp_x[1:30:end,1:30:end],pp_y[1:30:end,1:30:end],
angles="xy",scale=25);
```

```julia
download("http://ualberta.ca/~kstanton/files/kevin_0off_velocity.su","kevin_0off_velocity.su");
SegyToSeis("kevin_0off_velocity.su","kevins_model","su");
d,h = SeisRead("kevins_model");
```

```julia
coh,pp,res = SeisPWD(d,["w1"=>30,"w2"=>20,"dx"=>8,"dz"=>4]);
```

```julia
SeisPlot(d); 
Seismic.hold("true");
L = 1;
pp_x = cos((pp-90)*pi/180)*L
pp_y = sin((pp-90)*pi/180)*L
Seismic.quiver([0:30:350-1],[0:30:600-1],
pp_x[1:30:end,1:30:end], pp_y[1:30:end,1:30:end],
angles="xy",scale=25,color="purple");
```
