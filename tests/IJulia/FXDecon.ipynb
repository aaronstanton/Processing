```julia
using Seismic
```

```julia
d = SeisLinearEvents(["tau1"=>1.4,"nx1"=>500,"v1"=>-7000]);
d = d[:,:];
d = d + 0.2*randn(size(d));
```

```julia
d_fx = SeisFXDecon(d,["mu"=>0.05,"filter_length"=>20]);
```

```julia
Seismic.subplot(131)
SeisPlot(d,["vmin"=>-1,"vmax"=>1,"title"=>"Input"])
Seismic.subplot(132)
SeisPlot(d_fx,["vmin"=>-1,"vmax"=>1,"title"=>"FX Decon"])
Seismic.subplot(133)
SeisPlot(d - d_fx,["vmin"=>-1,"vmax"=>1,"title"=>"Difference"])
```
