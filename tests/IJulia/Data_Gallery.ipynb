```julia
SegyToSeis("/Users/astanton/Seismic/data/small_stack.su","small_stack","su",false);
d,h = SeisRead("small_stack");
nt = size(d,1);
dt = h[1].d1;
SegyToSeis("/Users/astanton/Seismic/data/wavelet_for_small_stack.su","wavelet_for_small_stack","su",false);
w,h_wav = SeisRead("wavelet_for_small_stack");
nt_wav = size(w,1);
plotpar = ["style"=>"color",
           "vmin"=>-1,"vmax"=>1,
           "aspect"=>"auto",
           "xlabel"=>"X","xunits"=>"","ox"=>0,"dx"=>1,
           "ylabel"=>"Time","yunits"=>"(seconds)","oy"=>0,"dy"=>dt,
           "cmap"=>"PuOr",
           "title"=>"small_stack.su"];

Seismic.subplot(121);
SeisPlot(d[:,:],plotpar);

Seismic.subplot(122);
w = [w[:]' zeros(nt-nt_wav)'];
Seismic.plot(w[:],[0:nt-1]*dt);
Seismic.title("wavelet_for_small_stack.su");
Seismic.suptitle("Stack and Wavelet");
ax = Seismic.gca();
ax[:invert_yaxis](); 
Seismic.plt.show();
```
